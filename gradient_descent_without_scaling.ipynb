```python
#Import Libiraries
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
#Generate Unscaled Data
m = 100

X1 = np.random.rand(m, 1) * 1000      # Large scale feature
X2 = np.random.rand(m, 1) * 1         # Small scale feature

X = np.hstack([X1, X2])
y = 5 * X1 + 2 * X2 + 10 + np.random.randn(m, 1) * 50
```

```python
#Add Bias Term
X_b = np.c_[np.ones((m, 1)), X]
```

```python
#Cost Function
def compute_cost(X, y, theta):
    m = len(y)
    predictions = X @ theta
    return (1/(2*m)) * np.sum((predictions - y)**2)
```

```python
#Gradient Descent
def gradient_descent(X, y, theta, lr, iterations):
    m = len(y)
    cost_history = []

    for _ in range(iterations):
        gradients = (1/m) * X.T @ (X @ theta - y)
        theta -= lr * gradients
        cost_history.append(compute_cost(X, y, theta))

    return theta, cost_history

theta = np.zeros((3,1))
theta, cost_history_unscaled = gradient_descent(X_b, y, theta, lr=0.0000001, iterations=100)
```

```python
#Plot Cost Curve (Without Scaling)
plt.figure(figsize=(6,4))
plt.plot(cost_history_unscaled)
plt.title("Cost vs Iterations (Without Scaling)")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.tight_layout()
plt.show()
```
